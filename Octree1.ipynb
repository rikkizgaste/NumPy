```python
import numpy as np
import laspy as lp
```

```python
point_cloud=lp.read('point cloud sample/2743_1234.las')
```

```python
point_format = point_cloud.point_format
```

```python
list(point_format.dimension_names)
```

```python
points = np.vstack((point_cloud.x, point_cloud.y, point_cloud.z)).transpose()
```

```python
points.size
```

```python
factor=10
decimated_points_random = points[::factor]
```

```python
decimated_points_random.size
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=(20, 20))
ax = plt.axes(projection='3d')
ax.scatter(decimated_points_random[:,0], decimated_points_random[:,1], decimated_points_random[:,2], s=0.01)
plt.show()
```
